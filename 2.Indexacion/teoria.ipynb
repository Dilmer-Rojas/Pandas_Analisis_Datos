### Veamos sobre ILOC y LOC en pandas

Se usan para seleccionar filas y columnas en un DataFrame

- `.iloc()` -> Selección por posiciones (índices númericos de filas y columnas)

- `.loc()` -> Selección por etiquetas (nombres de filas y columnas)

En resumen: `.iloc` selecciona las filas y columnas por índices
            y `.loc()` selecciona las filas y columnas por su nombre

#### .iloc()

```python
# .iloc() Selección de filas y columnas por índices
df.iloc[1] # Selecciona la fila 2 de posición 1
```

```python
df.iloc[:,1] # Selecciona la segunda columna `description`
```

```python
df.iloc[0, 2] # Obtiene el valor en la primera fila (0) y tercera columna (2)
```

```python
df.iloc[[0, 2], [0, 2]] 
```

#### .loc()

```python
df.loc[:, "country"] # Selecciona solo la columna 'country'
```

```python
df.loc[:, "price"] # Selecciona la columa precio
```

### Selección condicional

```python
# Vamoa a filtrar los datos donde country = Italy
df.country == "Italy"
```

```python
# Complementando con .loc() -> Filtramos solos las filas donde country es Italy
df.loc[df.country == "Italy"]
```

```python
# Tambien queremos filtrar los vinos mejores que el promedio (90) según su puntaje 
# El promedio se obtiene del puntaje de 80 a 100 -> 80 + 100 = 180 / 2 = 90
df.loc[(df.country == "Italy") & (df.points >= 90)]
```

```python
# Ahora queremos filtrar por country = Italy o points >= promedio (90)
df.loc[(df.country == 'Italy') | (df.points >= 90)]
```

```python
# Por ejemplo si queremos filtrar solo los vinos que son de Italy y France
# Hacemos uso del operador isin -> 'está en' -> está en una lista de valores
df.loc[df.country.isin(['Italy', 'France'])]
```

```python
# Queremos filtrar aquellos precios que son Null
df.loc[df.price.isnull()]
```

```python
# Filtrar aquellos precios que no son nulos
df.loc[df.price.notnull()]
```
