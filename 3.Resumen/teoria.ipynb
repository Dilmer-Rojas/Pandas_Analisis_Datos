## Resumen Funciones y Mapas

```python
import pandas as pd

ruta_file = "../1.Inicio/wine-reviews/versions/4/winemag-data-130k-v2.csv"
reviews = pd.read_csv(ruta_file, index_col=0)
reviews.head()
```

```python
# Función describe en Pandas
# Genera un resumen estadístico de un DataFrame o Serie en Pandas
# Por defecto describe() -> solo analiza columnas numéricas
# devolviendo
# count -> cantidad de valores no nulos
# mean -> media
# std -> desviación estandar
# min -> minimo
# max -> maximo
# 25% / 50% / 75% -> Percentiles

reviews.describe()
```

```python
# Para resumir de todas las columnas
reviews.describe(include="all")
```

```python
# Ver que tan bien le va a un vino con respecto al promedio de calificación
reviews.points.mean()
```

```python
# Para ver valores únicos usamos unique
reviews.country.unique()
```

```python
# Ver cuantas veces se repite un valor
reviews.country.value_counts()
```

```python
# Queremos centrar los datos alrededor de la media
reviews_points_mean = reviews.points.mean()
reviews.points.map(lambda p: p - reviews_points_mean)
```

```python
# Queremos aplicar esto a todas las columnas, usamos apply

def remean_points(row):
    row.points = row.points - reviews_points_mean
    return row

reviews.apply(remean_points, axis="columns")
```
